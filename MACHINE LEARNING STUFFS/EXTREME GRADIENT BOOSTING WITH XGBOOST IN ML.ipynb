# EXERCISE!

```python
import pandas as pd
import numpy as np

df=pd.read_csv("D:\DATASETS/mushrooms.csv")
df.head()
```

# adaptive boosting(exercise)!

```python
# importing stuffs for ADAPTIVE BOOSTING
from sklearn.ensemble import AdaBoostClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
#encoder= LabelEncoder()
#df = encoder.fit_transform(df)

for label in df:
    df[label] = LabelEncoder().fit(df[label]).transform(df[label])
```

```python
#CREATING ARRAYS FOR THE FEATURE AND THE TARGET VARIABLE
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
#SPLITING THE DATASET
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)

# INSTANTIATING THE DecisionTreeClassfier as: "treeclass"g
# INSTANTIATING THE AdaBoostClassifier as: "Adaboost"

#treeclass = DecisionTreeClassifier(criterion='entropy', max_depth=1)
AdaBoost = AdaBoostClassifier( n_estimators=400, learning_rate=1) #base_estimator=treeclass IS NOT NECESSARY

#TRAINING THE DATASET
AdaBoost.fit(x_train,y_train)
```

```python
#Evaluate the accuracy of the model
y_pred = AdaBoost.predict(x_test)
predictions = metrics.accuracy_score(y_test, y_pred)
#Calculating the accuracy in percentage
print('The accuracy is: ', predictions)
```

# EXERCISE FOR GRADIENT BOOSTING!

```python
import pandas as pd
import numpy as np

df=pd.read_csv("D:\DATASETS/mushrooms.csv")
df.head()
```

```python
# importing stuffs for GRADIENT BOOSTING
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
for label in df:
    df[label] = LabelEncoder().fit(df[label]).transform(df[label])
```

```python
#CREATING ARRAYS FOR THE FEATURE AND THE TARGET VARIABLE
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
#SPLITING THE DATASET
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)

# INSTANTIATING THE DecisionTreeClassfier as: "treeclass"g
# INSTANTIATING THE GradientBoostingClassifier as: "gradientboosting"

#treeclass = DecisionTreeClassifier(criterion='entropy', max_depth=1)
gradientboosting = GradientBoostingClassifier(n_estimators=400, learning_rate=1) #base_estimator=treeclass IS NOT NECESSARY

#TRAINING THE DATASET
gradientboosting.fit(x_train,y_train)
```

```python
#Evaluate the accuracy of the model
y_pred = gradientboosting.predict(x_test)
predictions = metrics.accuracy_score(y_test, y_pred)
#Calculating the accuracy in percentage
print('The accuracy is: ', predictions)
```

# EXERCISE FOR XG BOOSTING!

```python
import pandas as pd
import numpy as np

df=pd.read_csv("D:\DATASETS/mushrooms.csv")
df.head()
```

```python
# importing stuffs for XG BOOSTING
from xgboost import XGBClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier

from sklearn.model_selection import train_test_split
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
```

```python
for label in df:
    df[label] = LabelEncoder().fit(df[label]).transform(df[label])
```

```python
#CREATING ARRAYS FOR THE FEATURE AND THE TARGET VARIABLE
x = df.drop(['target'], axis=1)
y = df['target']
```

```python
#SPLITING THE DATASET
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,random_state=42)

# INSTANTIATING THE DecisionTreeClassfier as: "treeclass"
# INSTANTIATING THE GradientBoostingClassifier as: "xgboosting"

#treeclass = DecisionTreeClassifier(criterion='entropy', max_depth=1)
xgboosting = XGBClassifier(n_estimators=400, learning_rate=1) #base_estimator=treeclass IS NOT NECESSARY

#TRAINING THE DATASET
xgboosting.fit(x_train,y_train)
```

```python
#Evaluate the accuracy of the model
y_pred = xgboosting.predict(x_test)
predictions = metrics.accuracy_score(y_test, y_pred)
#Calculating the accuracy in percentage
print('The accuracy is: ', predictions)
```
