```python
import seaborn as sns
import pandas as pd
tips=sns.load_dataset("tips")
tips.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x="total_bill",y="tip",data=tips)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.scatterplot(x="total_bill",y="tip",data=tips,
                hue="smoker",hue_order=["No","Yes"])
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
hue_colors={"Yes":"m", #m=magenta
            "No":"c"}  #c=cyan
sns.scatterplot(x="total_bill",y="tip",data=tips,
                hue="smoker",
                palette=hue_colors)
plt.show()
```

```python
sns.countplot(x="smoker",data=tips,
             hue="sex")
```

```python
#BASICALLY CATPLOT/RELPLOT IS USED IN THE SITUATIO....WHEN WE ARE GONNA USE A "FACETGRID"...i.e.,LIKE A BUNCH OF SUBPLOTS...
#I MEAN IT HAS A BUNCH OF SUBPLOTS IN IT...WHEREAS IN "AXESSUBPLOT" ONLY A SINGLE SUBPLOT CAN BE MADE
#FACETGRID EXAMPLE- CATPLOT,RELPLOT(WE USE "KIND" COMMAND TO MENTION THE PLOT NAME...LIKE-BOX,SCATTER,COUNT...ETC)
#SUBPLOT EXAMPLE- WE CAN DIRECTLY USE ALL THESE "COUNTPLOT,BOXPLOT AND SCATTERPLOT)
g=sns.catplot(x="total_bill",y="tip",data=tips,kind="box")
g.fig.suptitle("new title",y=1.03)
```

```python
g=sns.relplot(x="total_bill",y="tip",  #relplot can also be replaced with catplot
              data=tips,
              kind="scatter")
type_of_g=type(g)
print(type_of_g)
```

```python
g=sns.relplot(x="total_bill",y="tip",data=tips,col="smoker",kind="line") #basically "col" is used to get different subplots  
g.fig.suptitle("TITTLE",y=1.05)
```
