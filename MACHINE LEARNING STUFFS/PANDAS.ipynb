```python
import pandas as pd
import numpy as np
dict1={"date":(1,2,3,4,5,6),
                "month":('jan','feb','mar','apr','may','june')}
df=pd.DataFrame(dict1)
dict2={"month":('july','aug','sep','oct','nov','dec'),
       "date":(7,8,9,10,11,12),}
df1=pd.DataFrame(dict2)
```

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(2,1)
ax[0].plot(df["date"],df["month"],marker="o",linestyle="None",color="r")
ax[1].plot(df1["month"],df1["date"],marker="v",linestyle="--",color="y")
ax[0].set_xlabel("time(months)")
ax[0].set_ylabel("date(months)")
ax[0].set_title("my 1st plot")
ax[1].set_xlabel("time(months)")
ax[1].set_ylabel("date(months)")
ax[1].set_title("my 1st plot")
plt.show()
```
