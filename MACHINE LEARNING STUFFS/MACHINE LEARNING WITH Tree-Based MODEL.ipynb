# 1) WHAT IS DECISION TREE CLASSIFICATION?


THIS IS WHAT A DECISION TREE IS!
IT'S JUST LIKE A JOURNEY OF QUESTIONS AND ANSWERS FROM "THE ROOT" TO THE "LEAVES"!

# NOTE-
    
    WHEN WE ARE IN NEED TO REPRESENT A "ROOT NODE"....WE SHOULD GO FOR THE MOST SUITABLE ONE!
FOR EXAMPLE-

THEN WE WOULD GO FOR CHECKING THE POSSIBILITIES OF THE THING....WHICH CAUSES HEART DISEASE THE "MOST"!
BY COMPARING THEM SEPERATELY WITH DIFFERENT LITTLE DECISION TRESS!


# BUT THEN MANY TIMES WE FACE THIS PROBLEM WHERE WE WOULD GET MANY "IMPURE" STUFFS


THE POPULAR AND THE EASY WAY TO FIND THE IMPURITIES IS
# GINI! (method)

# for example-

    HERE WE ARE EXAMINING THE GINI IMPURITY OF THE CHEST PAIN!
note- at 1st we are calculating the gini impurity of the "1st leaf node"(left one!)    


# AND NOW WE WOULD MOVE TO THE 2ND LEAF NODE(right one!)


# AND NOW WE ARE GONNA FINE OUT THE TOTAL AMOUNT OF 
# "GINI" IMPURITY!....COZ


# calculating the weighted average of the leaf node impurities


# HERE WE HAVE THE GINI IMPURITY FOR ALL THREE STUFFS

# As "Good blood circulation" has the least gini impurity...we should represent it as our "ROOT NODE"

AND BY FOLLOWING THIS....WE WOULD GO FOR "BLOCKED ARTERIES" AND THEN WE WOULD SWITCH TO CHEST PAIN!
AND BY THAT WAY...WE'VE CREATED OUR "DESICION TREE"

```python
import numpy as np
import pandas as pd

df=pd.read_csv("D:\DATASETS/iris.csv")
df.head()
```

```python
df.isna().sum()
```

```python
#CREATING ARRAYS FOR THE FEATURE AND THE TARGET VARIABLE
x= df.drop("Species",axis=1)
y= df["Species"]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

#SPLITING THE DATASET
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.3,stratify=y,random_state=42)

# INSTANTIATING THE DecisionTreeClassfier as: "treeclass"
treeclass= DecisionTreeClassifier()

#TRAINING THE DATASET
treeclass.fit(x_train,y_train)
```

# EVALUATE

```python
treeclass.score(x_test,y_test)
```

# EVALUATE

```python
treereg.score(x_test,y_test)
```

# 3)What is Random Forest?

GENERALLY "RANDOM FORESTS" ARE BUILT FROM "DECISION TREES"!

WKT,DECISION TREES ARE EASY TO BUILD AND USE...BUT IN PRACTICE THEY ARE NOT SO GREAT!



# SO IN ORDER TO CREATE A RANDOM FOREST...WE HAVE TO CREATE A "BOOT STRAPPED" DATASET!

    WHAT IS BOOTSTRAPPING?
    Bootstrapping is a statistical procedure that resamples a single dataset to create many simulated samples. 
    This process allows you to calculate standard errors, construct confidence intervals 
    and perform hypothesis testing for numerous types of sample statistics.

# TO CREATE A RANDOM FOREST...WE SHOULD CREATE N NUMBER OF NEW DECISION TREES (WITH BOOT STRAPPED DATASETS!)


# AS WE MADE A HELL LOT OF NEW DECISION TREES WITH BOOTSTRAPPED DATASET....WE'VE GOT VARIETIES OF OPTIONS


# SO HOW DO WE USE THIS "RANDOM FOREST"?

WELL THE ANSWER IS SO SIMPLE....WE JUST COLLECT THE DATA AND RUN IT DOWN IN EVERY NEWLY CREATED "DECISION TREES"(WITH BOOTSTRAPPED DATASET!).....AFTER COMPLETING THIS PROCESS...WE CHECK THE VOTES LIKE....WHICH RESULTS THE MOST!
